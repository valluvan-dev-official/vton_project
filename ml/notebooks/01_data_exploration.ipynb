# 01 — Data Exploration

Explore the training pairs collected by the API.

```python
import json
from pathlib import Path
import matplotlib.pyplot as plt
from PIL import Image

PAIRS_ROOT = Path('../../storage/training_pairs')
pairs = list(PAIRS_ROOT.glob('*/meta.json'))
print(f'Total pairs: {len(pairs)}')
```

```python
scores = [json.loads(p.read_text())['ssim_score'] for p in pairs]
plt.hist(scores, bins=20)
plt.xlabel('SSIM Score')
plt.ylabel('Count')
plt.title('Distribution of SSIM scores in training pairs')
plt.axvline(0.65, color='red', linestyle='--', label='Threshold (0.65)')
plt.legend()
plt.show()
```

```python
# Visualise a few pairs
for meta_path in pairs[:3]:
    meta = json.loads(meta_path.read_text())
    d = meta_path.parent
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))
    for ax, name in zip(axes, ['person', 'garment', 'result']):
        img_path = next(d.glob(f'{name}.*'), None)
        if img_path:
            ax.imshow(Image.open(img_path))
        ax.set_title(name)
        ax.axis('off')
    plt.suptitle(f"job_id={meta['job_id']}  ssim={meta['ssim_score']:.3f}")
    plt.tight_layout()
    plt.show()
```
