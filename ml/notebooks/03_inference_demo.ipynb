# 03 — Inference Demo

Load a trained checkpoint and run inference on an image pair.

```python
import sys
sys.path.insert(0, '..')

from src.inference.infer import VTONInference
import matplotlib.pyplot as plt
from PIL import Image
```

```python
CHECKPOINT = '../checkpoints/best.pt'   # adjust path
PERSON_PATH = '../data/raw/person/sample.jpg'
GARMENT_PATH = '../data/raw/garment/sample.jpg'
OUTPUT_PATH = '/tmp/vton_demo_output.jpg'

engine = VTONInference(CHECKPOINT)
engine.run(PERSON_PATH, GARMENT_PATH, OUTPUT_PATH)
print('Saved to', OUTPUT_PATH)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for ax, (title, path) in zip(axes, [('Person', PERSON_PATH), ('Garment', GARMENT_PATH), ('Result', OUTPUT_PATH)]):
    ax.imshow(Image.open(path))
    ax.set_title(title)
    ax.axis('off')
plt.tight_layout()
plt.show()
```
